# 切割影片畫面

```python
#https://blog.csdn.net/weixin_43202566/article/details/86528845
import os
import cv2

filepath = './images/'
if not os.path.exists(filepath):
    os.makedirs(filepath)

vidcap = cv2.VideoCapture('test.mp4')
success, image = vidcap.read()
fps= int(vidcap.get(cv2.CAP_PROP_FPS))
count = 0

while success:
    if count % fps ==0:
        cv2.imwrite(filepath+"image_%d.jpg" % int(count/fps), image)
    print('Process %dth seconds'% int(count/fps),success)
    success, image = vidcap.read()
    count += 1
```

# 將畫面進行處理

```python
#https://blog.csdn.net/lr94V587/article/details/127629669
#https://blog.csdn.net/m0_46296922/article/details/109902075

#https://developer.nvidia.com/cuda-toolkit-archive
#https://medium.com/ching-i/win10-%E5%AE%89%E8%A3%9D-cuda-cudnn-%E6%95%99%E5%AD%B8-c617b3b76deb

import numpy as np 
import cv2
import torch

print(torch.__version__)
print(torch.cuda.is_available())

size=(1280,720)
videowrite=cv2.VideoWriter(r'.\merge.mp4',-1,20,size)
img_array=[]
str=""
i=0
for filename in [r'.\images\image_{0}.jpg'.format(i) for i in range(50)]: #讀取圖片
    img=cv2.imread(filename)
    
    CONTENT_PATH = '.'+filename
    #print(CONTENT_PATH)
    if img is None:
        #print(filename +' is error !!')
        continue         
    
    f = open("settings.py", "w")

    # 寫入資料
    f.write("import torch\n")
    f.write("DEVICE = torch.device('cuda')\n")
    f.write("SIZE = 512\n")
    f.write("EPOCHS = 300\n")
    f.write("STYLE_PATH = './input/pasta.jpg'\n")
    f.write("STYLE_WEIGHT = 1000000\n")
    #f.write("CONTENT_PATH = '../input/liu.jpg'")
    str="CONTENT_PATH = '"+filename+"'\n"
    f.write(str)
    f.write("CONTENT_WEIGHT = 1\n")
    str="OUTPUT_PATH = './output/output_image_{0}.jpg'\n".format(i)
    f.write(str)
    
    f.close() # 關閉檔案
    
    %run main.py
    
for filename in [r'.\output\output_image_{0}.jpg'.format(i) for i in range(50)]: #讀取圖片
    img=cv2.imread(filename)
    
    #print(CONTENT_PATH)
    if img is None:
        #print(filename +' is error !!')
        continue   
    img_array.append()


for i in img_array:
    videowrite.write(i)
videowrite.release()
print('merge img end')
```

# 合併處理後的畫面和音檔


```python
import numpy as np 
import cv2
import torch

size=(1280,720)
videowrite=cv2.VideoWriter(r'.\merge.mp4',-1,20,size)
img_array=[]
for filename in [r'.\output\output_image_{0}.jpg'.format(i) for i in range(50)]: #讀取圖片
    img=cv2.imread(filename)
    if img is None:
        print(filename +' is error !!')
        continue   
    img_array.append(img)


for i in img_array:
    videowrite.write(i)
videowrite.release()
print('merge img end')
```
